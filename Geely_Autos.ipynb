**UNIVATRIATE ANALYSIS**

```python
#Univatriate distribution of the price feature
sb.histplot(df['price']);
```

- We observed that the distribution of the car feature is positively skewed.
- A greater number of cars from our data fall within the price range of 5,000 - 20,000

```python
#Univatriate distribution of the car
sb.countplot(x='fueltype',data = df);
```

- From the dataset it shows that most of the cars run on gas than diesel


```python
#Univatriate distribution of the carbody
sb.countplot(x='carbody',data = df);
```

- Observation shows that most of the cars are Sedan carbody followed by hatchback cars

```python
sb.countplot(x='enginetype',data =df);
```

- We observe from the dataset it shows that most runs on OHC engine


```python
#relationship between the price of the cars and the horsepower
sb.jointplot(x='horsepower',y='price',data = df);
```

```python
sb.scatterplot(x='horsepower',y='price',data = df);
```

- A positive correlation between horsepower feature and the dependant variable
- The horsepower feature is fairly symetrical with a trace of some outliers

```python
#relationship between the price of the cars and the engine location
sb.boxplot(x='enginelocation',y='price',data=df);
```

- From observation the data shows that on average less expensive cars with engine location in the front. Also, the data show presence of outliers
- Observation detects that all cars with engine location at the rear are on average more expensive than cars with engine location at the front

```python
sb.barplot(x='enginelocation',y='enginesize',data=df);
```

- Observed from the dataset that rear engine cars tend to be larger in size than front engine cars

```python
#relationship between engine type and highwaympg
sb.stripplot(x='enginetype',y='highwaympg',data=df);
```

- From observation the OHC engine on average has a higher mpg 

```python
sb.boxplot(x='fuelsystem',y='price',data = df);
```

```python
#relationship between car body and price
sb.boxplot(x='carbody',y='price',data =df);
```

- Observed that an average hardtop and convertible carbody are more expensive to Sedan carbody and wagon
- Also, from the dataset the least expensive carbody is hatchback

**MULTIVARIATE ANALYSIS**

```python
#relationship between 3 dataset
sb.boxplot(x='carbody',y='price', hue = 'enginelocation',data =df);
```

```python
sb.boxplot(x='carbody',y='price', hue = 'enginetype',data =df);
```

```python
sb.boxplot(x='fuelsystem',y='price', hue = 'enginetype',data =df);
```

```python
var = ['symboling','wheelbase','carlength','carwidth','stroke','citympg','compressionratio','price']
sb.pairplot(df[var]);
```

```python
plt.figure(figsize = (10,7))
sb.heatmap(df[var].corr(),annot = True);
```

```python
g=sb.pairplot(df[var],kind='reg',plot_kws ={'line_kws': {'color':'red'},'scatter_kws':{'alpha': 0.4}})
g.fig.set_size_inches(12,12)
g.fig.suptitle('Multivariate Analysis');
```
